```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from neuralforecast.models import TimesNet
from neuralforecast import NeuralForecast
from neuralforecast.losses.pytorch import MAE

def run_strategy(df_h_url, df_m_url):
    # 데이터 로드
    df_h = pd.read_csv(df_h_url)
    df_h = df_h.reset_index(drop=True)
    df_h_date = df_h['Datetime'][12:].reset_index(drop=True)
    df_h = df_h.rename(columns={"Datetime": "ds", "Close": "y"})
    df_h['unique_id'] = 'BE'
    df_h['ds'] = np.arange(len(df_h))

    # 모델 만들기
    horizon = 30
    models = [TimesNet(
        h=horizon,
        input_size=10,
        futr_exog_list=['Open', 'High', 'Low', 'Volume'],
        hist_exog_list=None,
        stat_exog_list=None,
        exclude_insample_y=False,
        hidden_size=64,
        dropout=0.1,
        conv_hidden_size=64,
        top_k=4,
        num_kernels=3,
        encoder_layers=2,
        loss=MAE(),
        valid_loss=None,
        max_steps=40,
        learning_rate=1e-4,
        num_lr_decays=-1,
        early_stop_patience_steps=-1,
        val_check_steps=100,
        batch_size=32,
        valid_batch_size=None,
        windows_batch_size=64,
        inference_windows_batch_size=256,
        start_padding_enabled=False,
        scaler_type='minmax',
        random_seed=1,
        num_workers_loader=0,
        drop_last_loader=False,
        optimizer=None
    )]

    nf = NeuralForecast(models=models, freq=1, local_scaler_type='minmax')
    nf.fit(df=df_h)

    # Cross Validation
    preds_df = nf.cross_validation(df=df_h, step_size=horizon, n_windows=1)
    preds_df = preds_df.reset_index()

    # 예측 결과를 실제 데이터와 결합
    real_df_h = df_h[-30:].reset_index(drop=True)
    real_df_h['TimesNet_pred'] = preds_df['TimesNet']

    # pred_signal 열 추가 및 값 입력
    real_df_h['pred_signal'] = ''
    for i in range(len(real_df_h) - 1):
        if real_df_h['TimesNet_pred'][i] < real_df_h['TimesNet_pred'][i + 1]:
            real_df_h.at[i, 'pred_signal'] = True
        else:
            real_df_h.at[i, 'pred_signal'] = False
    real_df_h.at[len(real_df_h) - 1, 'pred_signal'] = 'N/A'

    real_df_h['date'] = df_h_date

    # 두 번째 데이터 로드 및 pred_signal 결합
    df_m = pd.read_csv(df_m_url)
    df_m = df_m.reset_index(drop=True)
    df_m['pred_signal'] = ''
    for i in range(len(df_m)):
        for j in range(len(real_df_h)):
            if df_m['Datetime'][i][:10] == real_df_h['date'][j][:10]:
                df_m['pred_signal'][i] = real_df_h['pred_signal'][j]
            else:
                pass
    df_m['pred_signal'].replace('N/A', True, inplace=True)
    df_m['Datetime'] = pd.to_datetime(df_m['Datetime'])

    def apply_strategy(df_m, k):
        df_m['VBS_buy_signal'] = False
        for i in range(1, len(df_m)):
            if df_m['Close'].iloc[i] > df_m['Open'].iloc[i] + (df_m['High'].iloc[i-1] - df_m['Low'].iloc[i-1]) * k:
                df_m.at[df_m.index[i], 'VBS_buy_signal'] = True
            if df_m['Datetime'].iloc[i].minute == 0:
                df_m.at[df_m.index[i], 'VBS_buy_signal'] = False

        df_m['position'] = 0
        df_m['returns'] = 0.0
        df_m['cumulative_returns'] = 0.0
        df_m['trade_count'] = 0

        # 매수 및 매도 신호에 따라 포지션 관리 및 수익률 계산
        in_position = False
        buy_price = 0.0
        total_trades = 0
        for i in range(1, len(df_m)):
            if not in_position and df_m['pred_signal'].iloc[i] and df_m['VBS_buy_signal'].iloc[i]:
                df_m.at[df_m.index[i], 'position'] = 1
                buy_price = df_m['Open'].iloc[i]
                in_position = True
            elif in_position and df_m['Datetime'].iloc[i].minute == 0:
                df_m.at[df_m.index[i], 'position'] = 0
                sell_price = df_m['Close'].iloc[i]
                df_m.at[df_m.index[i], 'returns'] = (sell_price - buy_price) / buy_price
                df_m.at[df_m.index[i], 'trade_count'] = 1
                total_trades += 1
                in_position = False
            else:
                df_m.at[df_m.index[i], 'returns'] = 0.0

        df_m['cumulative_returns'] = (1 + df_m['returns'] - df_m['trade_count'] * 0.0003).cumprod() - 1
        return df_m['cumulative_returns'].iloc[-1], total_trades

    best_k = 0
    best_return = -np.inf
    total_trades_best_k = 0
    for k in np.linspace(0, 1, 21):
        current_return, total_trades = apply_strategy(df_m.copy(), k)
        if current_return > best_return:
            best_return = current_return
            best_k = k
            total_trades_best_k = total_trades

    print("Best k:", best_k)
    print("Best Return:", best_return)
    print("Total Trades:", total_trades_best_k)

    df_m_final = df_m.copy()
    apply_strategy(df_m_final, best_k)

    return df_m_final, best_k, total_trades_best_k

def plot_signals(df_m_final, best_k, total_trades):
    plt.figure(figsize=(14, 7))
    plt.plot(df_m_final['Datetime'], df_m_final['Close'], label='Close Price')
    buy_signals = df_m_final[(df_m_final['position'] == 1)]
    plt.plot(buy_signals['Datetime'], buy_signals['Open'], '^', markersize=10, color='g', label='Buy Signal')

    plt.xlabel('Date')
    plt.ylabel('Close Price')
    cumulative_return = df_m_final['cumulative_returns'].iloc[-1]
    plt.title(f'Close Price with Buy Signals (k={best_k.round(1)}, Cumulative Returns={(cumulative_return * 100).round(4)}%, Total Trades={total_trades})')
    plt.legend()
    plt.show()

    print(df_m_final[['Datetime', 'position', 'returns', 'cumulative_returns', 'trade_count']])

# 함수 실행
df_m_final, best_k, total_trades = run_strategy('https://raw.githubusercontent.com/CSID-DGU/2024-1-DSCD-LAST_DANCE-2/main/data/final_profit/weak_down_gaonchips_h.csv',
                                                'https://raw.githubusercontent.com/CSID-DGU/2024-1-DSCD-LAST_DANCE-2/main/data/final_profit/weak_down_gaonchips_m.csv')

# 신호 플롯
plot_signals(df_m_final, best_k, total_trades)
```
